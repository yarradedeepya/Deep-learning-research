```python
import numpy as np
import random

def fitness_function(x):
    return x[0]**2 + x[1]**2

def mto(population_size,active_food_sources,max_iter,bounds):
    dim = len(bounds)
    lb = [bounds[i][0] for i in range(dim)]
    ub = [bounds[i][1] for i in range(dim)]
    population = np.random.uniform(lb,ub, (population_size, len(bounds)))
    fitness = np.array([fitness_function(x) for x in population])
    best_index = np.argmax(fitness)
    best_solution=fitness[best_index]
    #print("best_soultion_value",best_solution)
    best_fitness = population[best_index]
    #print("best_soultion",best_fitness)
    sorted_indices = np.argsort(fitness)[::-1]
    a= population[sorted_indices]
    s= fitness[sorted_indices]
    '''print("population",population)
    print("fitness",fitness)
    print("best_index",best_index)
    print("best_fitness",best_fitness)
    print("sorted population",a)
    print("sorted solutions",s)'''
    NOS = (population_size//2)-1
    for i in range(max_iter):
            print("iteration",i)
            for j in range(0,NOS//2):
                    r=2*(np.round(np.random.rand(1,dim))-1)*(np.random.rand(1,dim))
                    #print(r)
                    a[0]=a[0]+0.5*r
                    print("TMT",a[0])
                    nf=fitness_function(a[0])
                    print("nf",nf)
                    if(nf>s[0]):
                        for k in range(0,NOS//2):
                            a[0]=a[0]+0.5*r      
                    else:
                        continue
            #Updating the Positions of FPCTs Candidate Solutions:
            for l in range(1,NOS):
                so=0
                print("l",l)
                for w in range(1,l):
                    print("w",w)
                    so=so+1/(l-w+1)*(best_fitness-a[l])
                   
                a[l]=a[l]+so
                ff=fitness_function(a[l])
                if(ff<s[l]):
                    a[l]=a[l]+0.5*r  
            #Updating the Positions of FCTs Candidate Solutions:      
            for m in range(population_size//2-1,population_size//2+2):
                po=0
                print("m",m)
                for c in range(m-NOS,m):
                    po=po+1/(m-c+1)*(best_fitness-a[m])
                a[m]=a[m]+po
               
            #Updating the Positions of LPCTs Candidate Solutions
            for z in range(population_size//2+2,population_size):
                print("z",z)
                go=0
                for g in range(z-NOS,population_size-NOS):
                    go=go+1/(z-g+1)*(best_fitness-a[g])
                a[z]=a[z]+go
            fitness = np.array([fitness_function(x) for x in a])
            best_index = np.argmax(fitness)
            best_solution=fitness[best_index]
            best_fitness = population[best_index]
            sorted_indices = np.argsort(fitness)[::-1]
            a= population[sorted_indices]
            s= fitness[sorted_indices]
            print("for iteration",i,"population postions are",a,"fitness values are",s)
   
    return best_solution, best_fitness 
    
```

```python
bounds = [(-5, 5), (-5, 5)]

# run the PSO algorithm to optimize the objective function
population_size = 6
max_iterations = 10
res1,res2=mto(population_size,0, max_iterations,bounds)
print("res1",res1)
print("res2",res2)
```
